```python
# Importação de Bibliotecas

import pandas as pd
```

```python
# Importando a base de dados e salvando como "df"

df = pd.read_csv("https://raw.githubusercontent.com/guilhermeonrails/data-jobs/refs/heads/main/salaries.csv")
```

```python
# Mostrando cabeçalho base de dados "df"

df.head()
```

```python
# Informações da base de dados "df"

df.info()
```

```python
# Dados descritivos básicos da base de dados "df" (Estatística Descritiva)

df.describe()
```

```python
# Exibir tamanho da base de dados "df" (shape é atributo e não parâmetro)

df.shape
```

```python
# Impressão tamanho da base de dados "df" usando estruturas do Pandas

linhas, colunas = df.shape[0], df.shape[1]

print("Número de linhas de df: ", linhas)
print("Número de colunas de df: ", colunas)
```

```python
# Lista dos nomes das colunas de "df"

df.columns
```

```python
# Traduzindo os nomes das colunas para português brasileiro

df.rename(columns={
    'work_year': 'ano',
    'experience_level': 'senioridade',
    'employment_type': 'contrato',
    'job_title': 'cargo',
    'salary': 'salario',
    'salary_currency': 'moeda',
    'salary_in_usd': 'usd',
    'employee_residence': 'residencia',
    'remote_ratio': 'remoto',
    'company_location': 'empresa',
    'company_size': 'tamanho_empresa'
}, inplace=True)

# Exibindo os nomes das colunas traduzidas
print(df.columns)

print("\n")

# Verificando resultado
df.head()
```

```python
df["senioridade"].value_counts()
```

```python
df["contrato"].value_counts()
```

```python
df["remoto"].value_counts()
```

```python
df["tamanho_empresa"].value_counts()
```

```python
# Renomeando categorias da coluna "senioridade" do dataframe "df"

senioridade = {
    'SE': 'Senior',
    'MI': 'Pleno',
    'EN': 'Junior',
    'EX': 'Executivo'
}

df['senioridade'] = df['senioridade'].replace(senioridade)

df['senioridade'].value_counts()
```

```python
# Renomeando categorias da coluna "contrato" do dataframe "df"

contrato = {
    'FT': 'Integral',
    'PT': 'Parcial',
    'CT': 'Contrato',
    'FL': 'Freelancer'
}

df['contrato'] = df['contrato'].replace(contrato)

df['contrato'].value_counts()
```

```python
# Renomeando categorias da coluna "tamanho_empresa" do dataframe "df"

tamanho_empresa = {
    'L': 'Grande',
    'S': 'Pequena',
    'M':	'Media'

}

df['tamanho_empresa'] = df['tamanho_empresa'].replace(tamanho_empresa)

df['tamanho_empresa'].value_counts()
```

```python
# Renomeando categorias da coluna "mapa_trabalho" do dataframe "df"
mapa_trabalho = {
    0: 'Presencial',
    100: 'Remoto',
    50: 'Hibrido'
}

df['remoto'] = df['remoto'].replace(mapa_trabalho)

df['remoto'].value_counts()
```
